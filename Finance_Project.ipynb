```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

We need to get data using pandas datareader. We will get stock information for the following Cell banks:

• Bank of America

• CitiGroup

• Goldman Sachs

• JPMorgan Chase

• Morgan Stanley

• Wells Fargo

```python
start= datetime.datetime(2006,1,1)
end= datetime.datetime(2016,1,1)
```

```python
BAC = data.DataReader('BAC','yahoo',start,end)
#citi group
CI = data.DataReader("C",'yahoo',start,end)
#Goldman Sachs
GS = data.DataReader("GS",'yahoo',start,end)
#JPMorgan
JPM = data.DataReader("JPM",'yahoo',start,end)
#Morgan Stanley
MS = data.DataReader("MS",'yahoo',start,end)
#Wells Fargo
WFC = data.DataReader("WFC",'yahoo',start,end)
```

```python
#MS
```

Create a list of the ticker symbols (as strings)  in alphabetical order.
Call this list: tickers

```python
tickers = ['BAC','CI','GS','JPM','MS','WFC']
```

Use pd.concat to concatenate the bank dataframes together to a single data frame called bank stocks. Set the keys argument equal to the tickers list. Also pay attention to what axis
you concatenate on.

```python
bank_stocks = pd.concat([BAC,CI,GS,JPM,MS,WFC],axis=1,keys=tickers)
```

```python
bank_stocks.head()
```

Set the column name levels (this is filled out for you):

```python
bank_stocks.columns.names = ['BanK Tickers', 'Stock Info']
```

```python
bank_stocks.head()
```

Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:

```python
returns = pd.DataFrame()
```

We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

# Moving Averages

Let's analyze the moving averages for these stocks in the year 2008.

Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008

```python
plt.figure(figsize=(12,4))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label='30 Day Avg')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC CLOSE')
plt.legend()
```

Create a heatmap of the correlation between the stocks Close Price.

```python
sns.heatmap(bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr(),annot=True)
```
